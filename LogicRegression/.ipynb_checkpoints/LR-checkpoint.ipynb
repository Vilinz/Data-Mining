```python
dataloader = Data.DataLoader(train_data, batch_size=200000, shuffle=True)
```

```python
class MyLogisticRegression(nn.Module):
    def __init__(self):
        super(MyLogisticRegression, self).__init__()
        self.linear = nn.Linear(32, 2)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        output = self.linear(x)
        output = self.sigmoid(output)
        return output
```

```python
model = MyLogisticRegression()

optimizer = torch.optim.SGD(model.parameters(), lr=0.001)
cross_fun = nn.CrossEntropyLoss()

model.train()
epochs = 1000
temp = Variable(torch.from_numpy(np.array(train_data)).float())
label = Variable(torch.from_numpy(np.array(label)).long())
label = label.squeeze()
for i in range(epochs):
    output = model(temp)
    loss = cross_fun(output, label)
    print(i, loss.item())
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
test = model(temp)
test = torch.max(test, 1)[1]
print(test.shape)
acc = 0
print(test.shape)
for i in range(len(train_data)):
   if test[i] == label[i]:
       acc += 1
print("acc: ", acc/len(train_data))
```
